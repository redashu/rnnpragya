```python
import  numpy as np
import  pandas as pd
import matplotlib.pyplot as plt
import keras
```

```python
# reading csv file 
df=pd.read_csv('trafficnew.csv',sep=',',header=None)
df.info()
```

```python
df.shape
```

```python
# visualize data
#all four wheelers vs buses
plt.xlabel('allfourwheelres')
plt.ylabel('bus')
plt.scatter(df[0],df[1],label="fourwheelers,buses")
plt.scatter(df[2],df[3],label='truck,minibus')
plt.grid()
plt.legend()
plt.show()
```

```python
#all four wheelers vs buses
plt.xlabel('allfourwheelres')
plt.ylabel('bus')
plt.scatter(df[2],df[3],label='truck,minibus')
plt.scatter(df[4],df[5],label='tractor,crane')
plt.grid()
plt.legend()
plt.show()
```

```python
#all four wheelers vs buses
plt.xlabel('allfourwheelres')
plt.ylabel('bus')
plt.scatter(df[0],df[1],label="fourwheelers,buses")
plt.scatter(df[4],df[5],label='tractor,crane')
plt.grid()
plt.legend()
plt.tight_layout()
plt.xticks(rotation=60)
plt.show()
```
